```
df = pd.read_csv('/content/autoimmune.csv')
```

```
abst = ' '.join(df['Abstract'])
```

```
model_bionlp = en_ner_bionlp13cg_md
model_bc5 = en_ner_bc5cdr_md
```

```
def display_entities(model,document):
    """ A function that returns a tuple of displacy image of named or unnamed word entities and
        a set of unique entities recognized based on scispacy model in use
        Args: 
            model: A pretrained model from spaCy or ScispaCy
            document: text data to be analysed"""
    nlp = model.load()
    doc = nlp(document)
    displacy_image = displacy.render(doc, jupyter=True,style='ent')
    #img = displacy.render(next(doc.sents), style='dep', jupyter=True)
    entity_and_label = set([(X.text, X.label_) for X in doc.ents])
    return  entity_and_label, displacy_image #, img
```

```
ent = display_entities(model = model_bc5,document = abst)
```

```
df_bc5  = pd.DataFrame(ent[0], columns = ['Entity','Label'])

df_bc5
```

```
ent = display_entities(model = model_bionlp,document = abst)
```

```
df_nlp  = pd.DataFrame(ent[0], columns = ['Entity','Label'])
df_nlp.head(50)
```

```
data = df_nlp.copy()
disease = df_bc5.copy()
disease['Label'] = disease['Label'].astype('str')
data['Label'] = data['Label'].astype('str')
data['Entity'] = data['Entity'].astype('str')
```

```
data = data[data['Label'] == "GENE_OR_GENE_PRODUCT"]
disease = disease[disease['Label'] == "DISEASE"]
```
